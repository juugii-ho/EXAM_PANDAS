# [ 데이터 시각화 맛보기 ]

## (1) 모듈 로딩

```python
import matplotlib.pyplot as plt
import pandas as pd
```

## <hr> (2) 데이터 준비

```python
# 2D 그래프를 그리기 위한 데이터 준비
X = range(1,100, 10)               # 0 ~ 99
Y = [value**2 for value in X]
# Y = []
# for d in X:
#     Y.append(d**2)
X,Y
```

## <hr> (3)  그래프 그리기

#### - [기본] 라인/선 그래프
    * 시간의 변화를 표현해주는 그래프
    * 연속적/순서 있는 데이터를 시각화할 경우 주로 사용됨

```python
plt.plot(X,Y)
```

```python
# plot(data) => y축 데이터로 사용
# plot(data1, data2) => x축에 data1, y축에 data2 데이터로 사용
# 반환값 : 라인/선 그래프 객체 리스트에 담겨서 반환됨

ret = plt.plot(X,Y)
type(ret), len(ret), type(ret[0])
```

```python
ret[0].get_xdata(), ret[0].get_ydata()
```

#### - 라인/선 그래프에서 형식 즉 포맷지정 => str 타입의 포맷 설정
    * '색상/형태/선스타일' ==> 'ro--' : red / circle / dot-line
    * 대표적으로 많이 사용되는 색상 : r - red, g - green, b - blue, k - black
    * 대표적으로 많이 사용되는 형태 : o - circle, ^- UpTriangle, <- LeftTriangle, >-RightTriangle, v-DownTriangle
    * 대표적으로 많이 사용되는 선 스타일 : - 실선, -- 점선, -.-. 실선 사이 점

- 라인/선 그래프에서 형식 즉 포맷지정 => str 타입의 포맷 설정
    * 색상-형태-선스타일 ==> 'ro--'(red, circle, dot-line)
    * 대표적으로 많이 사용되는 색상 : r - red, g - green, b - blue, c - cyan, m - magenta, y - yellow, k - black, w - white
    * 형태 : o - circle, ^ - triangle, s - square, v - pentagram, < - left triangle, > - right triangle, 8 - octagon, p - pentagram, * - star, h - hexagon, + - plus, x - x, D - diamond, d - thin_diamond, | - vline, _ - hline
    * 선스타일 : - - solid line, -- - dashed line, -. - dash-dot line, : - dot-dash line
    * 포맷 설정 방법 : plt.plot(x,y,'r--')

```python
ret=plt.plot(X, Y, 'ro')
```

```python
ret=plt.plot(X, Y, 'ro--')
```

```python
ret=plt.plot(X, Y, 'r^--')
```

```python
ret=plt.plot(X, Y, 'g*--')
```

```python
X2=[d //2 for d in X]
Y2=[d //2 for d in X]
ret=plt.plot(X, Y, 'g*--', X2, Y2, 'ro')
```
