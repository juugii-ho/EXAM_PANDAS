```python
import numpy as np
import matplotlib.pyplot as plt

X = np.linspace(-4, 4, 1024)
Y = .25 *(X+4.) * (X + 1.) * (X - 2.)

plt.title('A polynomial')
plt.plot(X,Y,c='k')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

X = np.linspace(-4, 4, 1024)
Y = .25 *(X+4.) * (X + 1.) * (X - 2.)

box = {
    'facecolor': '.75',
    'edgecolor': 'k',
    'boxstyle': 'round'
}

plt.text(-0.5, -0.25, 'Brackmard minimum', bbox=box)

plt.plot(X,Y,c='k')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

X = np.linspace(-4, 4, 1024)
Y = .25 *(X+4.) * (X + 1.) * (X - 2.)

plt.annotate('Brackmard minimum',
ha = 'center', va= 'bottom',
             xytext = (-1.5, 3.),
             xy = (0.75, -2.7),
             arrowprops = {'facecolor':'black', 'shrink':0.05})

plt.plot(X,Y,c='k')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

X = np.linspace(0, 6, 1024)
Y1 = np.sin(X)
Y2 = np.cos(X)

plt.xlabel('X')
plt.ylabel('Y')

plt.plot(X, Y1, c= 'k', lw=3., label = 'sin(X)')
plt.plot(X, Y2, c= '.5', lw=3., ls = '--', label = 'cos(X)')

plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

X = np.linspace(-4, 4, 1024)
Y = .25 *(X+4.) * (X + 1.) * (X - 2.)

plt.plot(X,Y,c = 'k')
plt.grid(True, lw = 2, ls ='--', c= '.75')

plt.show()
```
