## 텍스트 추가
- 그래프 내에 위치 지정 후 텍스트 추가 설정 => plt.text(시작x, 시작y, "출력할 텍스트")

```python
# 그래프 그리기
plotList=plt.plot(Y, 'ro--', Y2, 'g^')
plt.title("TEST", fontsize = 'xx-large', fontstyle= 'oblique', color = 'brown')

# x, y 축 레이블 
# x축 레이블 위치 loc='left' / 'center' / 'right'
# y축 레이블 위치 loc='bottom' / 'center' / 'top'
# 딕셔너리 값으로 다양한 설정값 기입 가능
plt.xlabel("값", loc='right', fontsize = 'large', color = 'purple')
plt.ylabel("결과", loc='top', fontsize = 'large', color = 'blue', rotation=0)

# x,y 축 눈금 설정
plt.xticks(ticks = [0, 2, 4, 6, 8], labels= ['A', 'B', 'C', 'D', 'E'], color='purple')
plt.yticks(color='blue')

# 범례 설정
plt.legend(labels=['빨강','초록'], loc='lower right')
#loc='upper left', 'upper right', 'lower left','lower right'

# 그래프 안에 텍스트 출력
plt.text(2, 2000, "Happy New Year", backgroundcolor= 'yellow', color = 'red')
box={'facecolor': '.75', 'edgecolor':'k', 'boxstyle':'round', 'alpha':0.2}
plt.text(2, 4000, "Happy New Year", bbox=box)
```
