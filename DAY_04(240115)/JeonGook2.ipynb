```python
# [기본] 모든 행에 1개의 값이라도 NaN이면 삭제
Korea2 = Korea.dropna()
print(f'Korea.shape {Korea.shape}, Korea2.shape{Korea2.shape}')
Korea2
```

```python
# [설정] 행에 모든 데이터 즉, 값이 NA면 삭제
Korea3 = Korea.dropna(how='all')
print(f'Korea.shape {Korea.shape}, Korea3.shape{Korea3.shape}')
Korea3
```

```python
# [설정] 특정 컬럼의 NA만 체크 후 삭제하는 방법
Korea4 = Korea.dropna(subset='1970')
print(f'Korea.shape {Korea.shape}, Korea4.shape{Korea4.shape}')
Korea4
```

```python
# [설정] 특정 컬럼의 NA만 체크 후 삭제하는 방법
Korea4 = Korea.dropna(subset='1970')
print(f'Korea.shape {Korea.shape}, Korea4.shape{Korea4.shape}')
Korea4
```

```python
# [설정] 특정 컬럼의 NA만 체크 후 삭제하는 방법
# thresh는 최소 기준 - how와 함께 못씀

Korea5 = Korea.replace('-', 0)
Korea5
Korea5_1 = Korea5.dropna(thresh=10, axis='index')
Korea5_1
```
