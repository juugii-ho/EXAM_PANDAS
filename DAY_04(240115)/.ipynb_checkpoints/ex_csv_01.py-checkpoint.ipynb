# [ CSV 데이터 파일 다루기 ]

## <hr> (1) 모듈 로딩

```python
import pandas as pd
```

## <hr> (2) 데이터 준비

- 파일 이름 => 경로 + 파이령
- 경로 => 2가지 종류
    * 절대경로 : 드라이브부터 파일까지의 전체 경로 (ex: C:\Users\data\.txt)
    * 상대경로 : 현재 파일을 기준으로 경로를 설정
       - ./  : 현재 파일과 동일한 폴더위치에 존재 (ex: ./a.csv 또는 a.csv) 
       - ../ : '현재 파일이 존재하는 폴더'의 한 단계 상위에 존재하는 파일 (ex:../Data/b.csv)

```python
# 아래 두 문장은 같은 의미
file = './test.csv'
```

```python
file = 'test.csv'
```

```python
file2 = '../DATA/movies.csv'
```

## <hr> (5) read_csv 파일 매개변수 살펴보기

```python
pd.read_csv(file2, header=None)
```

### - header 파라미터 : DataFrme의 컬러명 설정하는 파라미터
    * 기본값 : csv 파일의 첫번째 라인의 데이터를 컬럼명으로 설정
    * None : csv 파일의 첫번째 라인이 컬럼명이 아님을 의미 => 컬럼명 미지정
      DataFrame(data, columns=['A', 'B'], index=[10,20,30])일 때
      DataFrame(data, columns=None      , index=[10,20,30])처럼 됨
    * 행번호 : csv 파일의 행번호 라인이 컬럼명으로 설정, 그 이전 행 번호 데이터는 사라짐

```python
pd.read_csv(file2) #header를 부여하지 않으면 자동으로 생김
```

```python
pd.read_csv(file2, header=5) 
#header를 부여하면 해당 행이 헤더가 되고 이전 행은 사라짐
```

### - index_col 파라미터 : 컬럼 중에서 행 인덱스로 설정, set_index와 동일 역할

```python
# 아래 두 식은 같은 값임
pd.read_csv(file2,index_col='Rank')
pd.read_csv(file2,index_col=0)
```

```python
# 멀티인덱스도 가능, 아래 두 식은 같은 값임
pd.read_csv(file2,index_col=['Rank','Title'])
pd.read_csv(file2,index_col=[0,1])
```

### - sep/delimiter : csv 파일에서 데이터 구분자 설정
    * 기본값 : 쉼표(,)

```python
pd.read_csv(file2, sep=';') #구분자가 잘못 설정되면 아래와 같이 됨
```

```python
pd.read_csv(file) # 첫 번째 줄이 컬럼명인지 항상 확인해야 함
```

```python
pd.read_csv(file, header=2, sep=';')
```

### - skiprows : csv 파일 상단에서 읽어 들이지 않을 행 갯수를 지정하기

```python
pd.read_csv(file, skiprows=3, header=None, sep=';')
```

### -skipfooter : csv 파일 하단에서 읽어 들이지 않을 행 갯수를 지정하기

```python
pd.read_csv(file, skiprows=3, header=None, sep=';', skipfooter=2, engine='python')
```

### -usecols : 컬럼에서 특정 컬럼만 선택해서 불러오기
    * usecols = [0,2, ...]
    * usecols = ['컬럼명1', ..., '컬럼명n']

```python
pd.read_csv(file, sep=';', skiprows=3, header=None, usecols=[0,2])
```

```python
# 아래 두 식은 같은 값임
pd.read_csv(file2, usecols=['Rank', 'Studio'])
pd.read_csv(file2, usecols=[0,2])
```
