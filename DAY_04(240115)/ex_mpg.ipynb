```python
horseSR = mpgDF.마력
for idx, data in enumerate(horseSR):
    if data =='?':
        print(horseSR[idx])
        horseSR[idx] = '0.0'
        print(horseSR[idx])
mpgDF.마력 = horseSR.astype('float64')
mpgDF.info(3)
```

## [ 데이터 원소값 변경 / 타입 변경 한꺼번에 진행 ]

```python
# 데이터 읽어오면서 바로 컬럼명 지정
mpgDF2 = pd.read_csv(file, 
                     header=None, 
                     sep='\s+', 
                     names=["mpg", "cylinders", "displacement", "horsepower", "weight", "acceleration", "model year", "origin", "car name"])
```

```python
#데이터 정보 확인
mpgDF2.info()
```

```python
# 원소값 '?' 변경해주는 메서드 => replace(old, new)
mpgDF2.horsepower.replace('?', '0.0', inplace=True)
mpgDF2
```

```python
# origin 컬러의 1- USA, 2- EU, 3- JPN
mpgDF2.origin.replace({1 :'USA', 2: 'EU', 3:'JPN'}, inplace=True)
mpgDF2
```

```python
# origin 컬럼의 값의 종류별 갯수 확인 => value_counts() / unique()
mpgDF2.origin.value_counts()
```

```python
# 고유값 즉 값의 종류와 갯수
print( mpgDF2.origin.unique(), mpgDF2.origin.nunique())
```

##### horsepower컬럼과 origin컬러의 데이터 타입 변경

```python
mpgDF2.horsepower = mpgDF2.horsepower.astype('float')
mpgDF2.origin = mpgDF2.origin.astype('category')
```

```python
mpgDF2.info()
```
